```python
import numpy as np
import matplotlib.pyplot as plt

D = 20
dimensions = [2**i for i in range(D)]

def l1_norm(x, y):
    return np.sum(np.abs(x - y))

def l2_norm(x, y):
    return np.sum(np.square(x - y))

l1_norm_avgs = np.zeros(D)
l1_norm_stds = np.zeros(D)

l2_norm_avgs = np.zeros(D)
l2_norm_stds = np.zeros(D)

for i, d in enumerate(dimensions):
    points = np.random.rand(d, 100)
    l1_distances = np.array([l1_norm(x, y) for x in points for y in points])
    l2_distances = np.array([l2_norm(x, y) for x in points for y in points])

    l1_norm_avgs[i] = np.mean(l1_distances)
    l1_norm_stds[i] = np.std(l1_distances)

    l2_norm_avgs[i] = np.mean(l2_distances)
    l2_norm_stds[i] = np.std(l2_distances)
```

```python
print(l2_norm_avgs)
```

```python
fig = plt.figure()

ax = fig.add_subplot(3, 1, 1)

ax.scatter(dimensions, l1_norm_avgs, marker='o', s=6, label='$\ell_1$ norm')
ax.plot(dimensions, l1_norm_avgs, alpha=0.6)
ax.scatter(dimensions, l2_norm_avgs, marker='o', s=6, label='$\ell_2$ norm')
ax.plot(dimensions, l2_norm_avgs, alpha=0.6)
ax.set_title('Average norms')
ax.set_xlabel('Dimension (log)')
ax.set_ylabel('Distance')

ax.set_xscale('log')
ax.legend()

ax = fig.add_subplot(2, 1, 2)

ax.scatter(dimensions, l1_norm_stds, marker='o', s=6, label='$\ell_1$ norm')
ax.plot(dimensions, l1_norm_stds, alpha=0.6)
ax.scatter(dimensions, l2_norm_stds, marker='o', s=6, label='$\ell_2$ norm')
ax.plot(dimensions, l2_norm_stds, alpha=0.6)
ax.set_title('Standard deviations of norms')
ax.set_xlabel('Dimension (log)')
ax.set_ylabel('Distance')

ax.set_xscale('log')
ax.legend()

fig.subplots_adjust(hspace=0.3)

plt.show()
```
